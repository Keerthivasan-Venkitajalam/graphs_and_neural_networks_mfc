```python
import csv

def extract_sentences(input_file, output_file):
    with open(input_file, 'r') as f_input, open(output_file, 'w', newline='') as f_output:
        csv_writer = csv.writer(f_output)
        csv_writer.writerow(['label', 'sentence'])  # Write header
        
        current_label = None
        nodes = []

        for line in f_input:
            line = line.strip()
            if not line:
                if current_label is not None:
                    sentence = ' '.join(nodes)
                    csv_writer.writerow([current_label, sentence])
                current_label = None
                nodes = []
            else:
                if line.startswith('n'):
                    _, _, node_text = line.split(' ', 2)
                    nodes.append(node_text)
                elif line.startswith('x'):
                    _, label = line.split(' ', 1)
                    current_label = label

        if current_label is not None:
            sentence = ' '.join(nodes)
            csv_writer.writerow([current_label, sentence])

# Usage example
extract_sentences("D:\\New folder\\Twitter-Graph\\tdats.nel", 'output.csv')
import csv

def parse_csv(file_path):
    graphs = []
    with open(file_path, 'r') as f:
        csv_reader = csv.reader(f)
        next(csv_reader)  # Skip header
        for row in csv_reader:
            label = row[0]
            sentence = row[1]
            graphs.append({
                'label': label,
                'sentence': sentence
            })
    return graphs

graphs = parse_csv('output.csv')
for i, graph in enumerate(graphs):
    if i == 10:
        break
    print('Label:', graph['label'])
    print('Sentence:', graph['sentence'])
    print()
```
